```python
plt.figure(figsize = (10, 5))
sns.boxplot(y= 'int.rate', x = 'not.fully.paid', data = df)
plt.title('Default by Interest Rate')
plt.savefig("loan_default_by_InterestRate.png")
plt.xlabel('Default')
plt.ylabel('Interest Rate')
plt.show()
```

## feature Selection

```python
df.head(1)
```

```python
df_encoded = pd.get_dummies(df, columns = ['purpose'], drop_first = True)
```

```python
df_encoded
```

```python
x = df_encoded.drop(['not.fully.paid'], axis = 1)
y = df['not.fully.paid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x_scaled = StandardScaler().fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter = 1000, solver = 'liblinear', class_weight = 'balanced')
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
results = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': y_pred
})
```

```python
results
```

```python
results['Predicted'].value_counts()
```

```python
results['Actual'].value_counts()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot = True, fmt = 'd', cmap = 'rocket')
plt.title('Relation between actual and predictions')
plt.xlabel('Predicted_Default')
plt.ylabel('Actual_Default')
plt.show()
```

```python
results = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': y_pred,
    'FICO': x_test[:, list(x.columns).index('fico')],  # example
    'Interest Rate': x_test[:, list(x.columns).index('int.rate')],
})
```

```python
results.to_csv("loan_predictions_results.csv", index=False)
```
